```python
import numpy as np
import pandas as pd
```

## 读取MySQL表格数据

```python
# !pip install pymysql
```

```python
# !pip install sqlalchemy
```

```python
import pymysql
from sqlalchemy import create_engine
```

```python
# 建立连接
db_info = {
    'user':'root',  
    'password':'',  
    'host':'localhost',  
    'database':'pandas_test'  # 这里我们事先指定了数据库，后续操作只需要表即可
}

# 推荐使用，可设置参数echo=True，会显示在加载数据库所执行的SQL语句
conn = create_engine('mysql+pymysql://%(user)s:%(password)s@%(host)s/%(database)s?charset=utf8' % db_info, 
                       encoding='utf-8')

# 不推荐
# conn = pymysql.connect(host="localhost", user="root", passwd="", db="pandas_test", charset='utf8')
```

```python
# 读取数据表格1
sql_1 = "select * from 成绩单1"
data_table_1 = pd.read_sql(sql_1, conn)
data_table_1
```

```python
# 读取数据表格2
sql_2 = "select * from 成绩单2"
data_table_2 = pd.read_sql(sql_2,conn)
data_table_2
```

## 数据处理★

### 表格拼接

```python
combined_data_table = pd.concat([data_table_1, data_table_2], axis=0, sort=True, ignore_index=True)
combined_data_table
```

### 去重

```python
combined_data_table
```

```python
# 按整条记录进行去重
combined_data_table.drop_duplicates(inplace=True)
combined_data_table
```

```python
# 按某一列进行去重(例如列举所有不同星座，每个星座只需要有一条记录代表)
combined_data_table.drop_duplicates("星座")
```

### 异常值处理

```python
# 查看成绩列是否有小于0或大于100的值
outliers_bool_sr = (combined_data_table["成绩"] < 0) | (combined_data_table["成绩"] > 100)
outliers_bool_sr
```

```python
# 异常值个数
sum(outliers_bool_sr)
```

```python
# 异常值的行标签
combined_data_table.loc[outliers_bool_sr].index
```

```python
# 异常值处理
combined_data_table.drop(labels=combined_data_table.loc[outliers_bool_sr].index, inplace=True)
combined_data_table
```

### 空值处理★

#### 检查缺失值（isnull和notnull）

```python
# 打印各元素是否是空值
combined_data_table.isnull()
```

```python
# 打印各元素是否非空值
combined_data_table.notnull()
```

```python
# 统计每一列有多少空值
combined_data_table.isnull().sum()
```

```python
combined_data_table.shape
```

```python
# 统计每一列有多少非空值
combined_data_table.notnull().sum()
```

```python
# 统计全为空值的列个数
sum(combined_data_table.notnull().sum() == 0)
```

#### 填充空值

```python
# 查看成绩为空的那条记录
combined_data_table.loc[combined_data_table["成绩"].isnull(), :]
```

```python
# 为该条记录填充语文的平均成绩
chinese_mean_score = combined_data_table.loc[combined_data_table["学科"]=="语文", "成绩"].mean()
combined_data_table["成绩"].fillna(chinese_mean_score, inplace=True)
combined_data_table
```

#### 删除全为空的列

```python
combined_data_table.dropna(axis=1, how='all', inplace=True)  # any
combined_data_table
```

#### 删除有空值的行

```python
combined_data_table.dropna(axis=0, how='any', inplace=True)
combined_data_table
```

### 数据标准化★

#### 离差标准化

```python
combined_data_table["离差标准化成绩"] = (combined_data_table["成绩"] - combined_data_table["成绩"].min()) \
                                    / (combined_data_table["成绩"].max() - combined_data_table["成绩"].min())

combined_data_table
```

#### 标准差标准化

```python
combined_data_table["标准差标准化成绩"] = (combined_data_table["成绩"] - combined_data_table["成绩"].mean()) \
                                    / combined_data_table["成绩"].std()

combined_data_table
```

### 哑变量处理（独热编码）★

```python
one_hot_features = pd.get_dummies(combined_data_table["星座"])
one_hot_features
```

```python
processed_data_table = pd.concat([combined_data_table, one_hot_features], axis=1)
processed_data_table
```

### 批量操作★

```python
# 每个元素进行相同操作
processed_data_table["星座"].apply(lambda x: x+"座")
```

```python
def func(string):
    return string+"座"

processed_data_table["星座"].apply(func)
```

```python
# 整列合并计算
processed_data_table.agg({"成绩": np.mean, "星座": 'mode'})
```

## 分组统计

```python
groupby_res = processed_data_table.groupby(by=["学科"], axis=0)
groupby_res
```

```python
# 对分组结果进行批量变换
groupby_res.apply(lambda x: x)
```

```python
# 分组统计均值
groupby_res.apply(lambda x: x.mean())
```

```python
# 对每个成绩进行对应学科的标准差标准化
groupby_res["成绩"].apply(lambda x: (x-x.mean())/ x.std() )
```

```python
# 分组差别化统计
groupby_res.agg({"姓名": 'count', "成绩": 'mean'})
```

## 输出MySQL数据表

```python
processed_data_table
```

```python
processed_data_table.to_sql('after_processed', conn, index=False, if_exists='append')
```
